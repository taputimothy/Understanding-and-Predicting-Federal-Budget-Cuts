```python
import pandas as pd
merged_df = pd.read_csv("merged_df.csv").drop_duplicates(subset=['fain_key'])
merged_df
```

```python
doge_grant = pd.read_csv("doge-scrape/data/doge-grant.csv")
doge_grant
```

```python
condensed = doge_grant[['fain', 'place_of_performance_county_name']].rename(columns={'fain': 'fain_key'})
condensed
```

```python
new_doge = pd.merge(merged_df, condensed, on = 'fain_key', how = 'left').drop_duplicates(subset=['fain_key'])
```

```python
r1_counties = ['DURHAM', 'WAKE','MECKLENBURG', 'ORANGE', 'PITT']
new_doge['County with R1 University?'] = new_doge['place_of_performance_county_name'].isin(r1_counties)
```
